# Importating Libraries


```
# Importing the libraries for analysis
import pandas as pd
```

# Reading the Data of our csv file


```
# we read the data from the csv file and create a dataframe(df)

electric_cars = pd.read_csv('/content/Autolib_dataset (2).csv',delimiter=",")
electric_cars.head(10)
```

# Accessing our Data sets info (rows columns etc)


```
electric_cars.info()
```

# Cleaning our data set


```
# We shall use 5 data cleaning procedures
# The first one is Validity
# We shall drop the columns unnecesary to our analysis
# Then we check our dataset's info again to confirm

electric_cars.drop(['Displayed comment','Geo point','Address','Cars'], axis=1, inplace =True)
electric_cars.info()

```

```
# Next we check our data for completeness
# We check for null values across all rows 

electric_cars.isnull()
```

```
# Do the same for columns

electric_cars.isnull().sum()

```

```
# Hence we now drop all those that had null values to confirm 
# completeness
electric_cars_dropped =electric_cars.dropna( axis=1)
electric_cars_dropped.isnull().sum()
```

```
# Next property is consistency 
# But in the case of our dataset there were no duplicates
# 
electric_cars_dropped[electric_cars_dropped.duplicated()]
clean_data = electric_cars_dropped.drop_duplicates()
```

```
# Finally we check for uniformity
# in years months days hours and minutes
clean_data['Date_time'] =pd.to_datetime(clean_data[['year','month','day','hour','minute']])
clean_data.head()
```

# Clean Data-set Importation


```
# We shall export the dataset using the function _csv
# as shown below

clean_data.to_csv('autolib__dataset_cleaned.csv')
```

# Problem Questions for analysis
 






 



```
#identifyng the city of paris only with its values in various columns
# 
clean_data_paris = clean_data[clean_data['City']== 'Paris']
clean_data_paris
```

```
# Finding the most frequent time an electic blue car 
# is picked up during the day in the month of april
blue_car = clean_data_paris['Bluecar counter'].groupby(clean_data['Date_time'])
blues_car_sum = blue_car.sum()
blues_car_sum.sort_values(ascending = False)
```

```
# Most frequent time for returning the cars
frequent_time = clean_data_paris['Bluecar counter'].groupby(clean_data['Date_time'])
frequent_time_count = frequent_time.count()
frequent_time_count.sort_values(ascending = False).head(1)
```

```
# Most frequently visited station
clean_data_paris[(clean_data_paris['Kind'] == 'STATION') & (clean_data_paris['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values( ascending = False)
```

```
#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?
clean_data_paris[(clean_data_paris['Kind'] == 'STATION') & (clean_data_paris['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False).head(1)
#Themost frequented station for blue cars does not match its postal code
```

```
# Utilib counter
# Most popular time to return the cars
frequent_time = clean_data_paris['Utilib counter'].groupby(clean_data['Date_time'])
frequent_time_count = frequent_time.count()
frequent_time_count.sort_values(ascending = False).head(1)
```

```
# Utilib counter 1.4
# Find the most popular time of the day for picking up a (Utilib) 
# in Paris during april 2018
blue_car = clean_data_paris['Utilib 1.4 counter'].groupby(clean_data['Date_time'])
blue_car_sum = blue_car.sum()
blue_car_sum.sort_values(ascending = False)
```

```
# Utilib counter
# Identify the most popular hour of the day for picking up a Utilib
#in the city of Paris over the month of April 2018
blue_car = clean_data_paris['Utilib counter'].groupby(clean_data['Date_time'])
blues_car_sum = blue_car.sum()
blues_car_sum.sort_values(ascending = False)
```
